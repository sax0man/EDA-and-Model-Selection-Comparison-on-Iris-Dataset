```python
def boxplot_iris(feature_names,dataset):
    i=1
    plt.figure(figsize=(11,9))
    for col in feature_names:
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both',left=True,top=False,right=False,bottom=True,labelleft=False,labeltop=False,labelright=False,labelbottom=False)
        dataset[col].plot(kind='box',subplots=True,sharex=False)
        plt.title(col)
        i+=1
    plt.show()
```

```python
boxplot_iris(iris.columns[:-1],iris)
```

```python
def histogram_iris(feature_names,dataset):
    i=1
    plt.figure(figsize=(11,9))
    for col in feature_names:
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both',left=True,top=False,right=False,bottom=True,labelleft=False,labeltop=False,labelright=False,labelbottom=False)
        dataset[col].plot(kind='hist',subplots=True,sharex=False)
        plt.title(col)
        i+=1
    plt.show()
```

```python
histogram_iris(iris.columns[:-1],iris)
```

```python
corr=iris.corr()
cmap=sns.diverging_palette(220,10,as_cmap=True)
plt.figure(figsize=(11,9))
sns.heatmap(corr,cmap=cmap,vmax=1.0,vmin=-1.0,center=0,square=True,linewidths=0.5,cbar_kws={'shrink':.5})
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.plot()
```

```python
def piechart_iris(feature_names,target,dataset):
    i=1
    plt.figure(figsize=(11,9))
    for colName in [target]:
        labels=[];sizes=[];
        df=dataset.groupby(colName).size()
        for key in df.keys():
            labels.append(key)
            sizes.append(df[key])
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both',left=True,top=False,right=False,bottom=True,labelleft=True,labeltop=True,labelright=False,labelbottom=False)
        plt.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=140)
        plt.axis('equal')
        i+=1
    plt.show()
```

```python
piechart_iris(iris.columns[:-1],iris.species,iris)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(iris.iloc[:,:-1],iris.iloc[:,-1],test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=None,min_samples_split=2,min_samples_leaf=1,
                            min_weight_fraction_leaf=0.0,max_features=None,random_state=42,max_leaf_nodes=None,
                            min_impurity_decrease=0.0,class_weight=None)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score,KFold
cv=KFold(n_splits=10,shuffle=True,random_state=42)
results=cross_val_score(model,X_train,Y_train,cv=cv)
fin_result=np.mean(results)
```

```python
for i,k in enumerate(results):
    print(f"{i}th cross validation score:{k}")
print(f"\n Final Cross Validation score: {fin_result}")
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
results=cross_val_score(model,X_train,Y_train,cv=cv)
for i,k in enumerate(results):
    print(f"{i}th stratified cross validation score: {k}")
```

```python
print(f"Final Stratified cross validation score is {np.mean(results)}")
```

```python
estimator=DecisionTreeClassifier()
from sklearn.model_selection import GridSearchCV
parameters={'max_depth':[4,6,8,10,12],
            'criterion':['gini','entropy'],
           'splitter':['best','random'],
           'min_weight_fraction_leaf':[0.0,0.1,0.2,0.3],
           'random_state':[7,23,42,78,142],
           'min_impurity_decrease':[0.0,0.05,0.1,0.2]}
model2=GridSearchCV(estimator=estimator,param_grid=parameters, cv=StratifiedKFold(10),verbose=1,n_jobs=-1,refit=True)
model2.fit(X_train,Y_train)
```
